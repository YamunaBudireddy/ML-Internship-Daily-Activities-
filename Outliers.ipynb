# IQR Method

```python
Q1=data.quantile(0.25)
Q3=data.quantile(0.75)
IQR=Q3-Q1
LB=Q1-1.5*IQR
UB=Q3+1.5*IQR
Outliers=data[(data< (LB))|(data>(UB))]
Outliers
```

# Z-Score Method

```python
import pandas as pd
data=pd.DataFrame({
    "Hours":[1,2,3,4,5,6,7,8,9,20]
})
data
```

```python
import pandas as pd
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
z_scores=np.abs(stats.zscore(data["Hours"]))
Outliers=data[z_scores>2]
Outliers
```

```python
stats.zscore(data["Hours"])
```

```python
plt.scatter(range(len(data)),data["Hours"])
plt.xlabel("Index")
plt.ylabel("Hours")
plt.title("Scatter plot Outliers")
plt.show()
```

```python
plt.hist(data["Hours"],bins=40)
plt.xlabel("Hours")
plt.ylabel("Frequency")
plt.title("Histogram Outliers")
plt.show()
```
