```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.DataFrame({
    "age":[18,22,25,30,35,40,45,50]
})
print(data)
```

```python
data["age_bin"]=pd.cut(data["age"],bins=3)
data
```

```python
plt.hist(data["age"],bins=4)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Equal-Width Binning using pd.cut()")
plt.show()
```

```python
plt.hist(data["age"],bins=3)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Equal-Width Binning using pd.cut()")
plt.show()
```

# Custom domain based Binning

```python
import pandas as pd
custom_bins=[17,25,35,50]
labels=["Young","Adult","Senior"]

bins = pd.cut(data2["age"], bins=custom_bins, labels=labels)
bins
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.hist(bins, bins=4)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Equal_Width Binning Using pd.cut()")
plt.show()
```
