# Creating Array

```python
a = np.array([1,2,3,4]) # one dimensional array
print(a.ndim)
```

```python
b = np.array([[1,2,3],[4,5,6]]) # two dimensional array
print(b.ndim)
```

```python
c = np.array([[[1,2],[3,4]],[[5,6],[7,8]]]) # multi dimensional array
print(c.ndim)
print(c)
```

```python
print(a.shape) #vector
```

```python
print(b.shape) # matrix
```

```python
print(c.shape) # cube
```

# coping and Viewing

```python
a = np.arange(10)
b = a[::2]
print(a)
print(b)
```

```python
a = np.arange(10)
b = a[::2]
print(a)
print(b)
```

```python
a = np.arange(10)
b = a[::2]
print(a)
print(b)
```

```python
b[0] = 10
print(a)
print(b) # to confirm share memory
```

```python
a = np.arange(10)
b = a[::2].copy() #force a copy
print(a)
print(b)
```

```python
a = np.arange(10)
b = a[::2].copy() #force a copy
print(a)
print(b)
```

```python
print(np.shares_memory(a,b))
```
